```python
# Day 4 Exercise
import pandas as pd
import numpy as np

# Load the messy dataset
df2 = pd.read_csv('data/messy.csv')

print("\nExercise 1")
# Print the data type of every column
print(df2.dtypes)

print("\nExercise 2")
# Convert object columns that should be numeric
df2['Peak'] = pd.to_numeric(df2['Peak'], errors='coerce')
df2['All Time Peak'] = pd.to_numeric(df2['All Time Peak'], errors='coerce')
print(df2[['Peak', 'All Time Peak']].dtypes)

print("\nExercise 3")
# Count missing values per column
print(df2.isnull().sum())

print("\nExercise 6")
# Drop rows where more than 2 values are missing — DO THIS BEFORE FILLING
df2 = df2.dropna(thresh=len(df2.columns) - 2)

print("\nExercise 4")
# Fill missing numeric values with column mean — AFTER dropping bad rows
numeric_cols = df2.select_dtypes(include='number').columns
for col in numeric_cols:
    df2[col] = df2[col].fillna(df2[col].mean())

print("\nExercise 5")
# Fill missing categorical values with most frequent value
categorical_cols = df2.select_dtypes(include='object').columns
for col in categorical_cols:
    df2[col] = df2[col].fillna(df2[col].mode()[0])

print("\nExercise 7")
# Confirm zero missing values
print(df2.isnull().sum())

print("\nCleaned dataset preview:")
print(df2.head(20))
```

```python
import pandas as pd

#Exercise 1: Load your cleaned heart.csv and confirm zero nulls and correct dtypes in one cell
#Exercise 2: Answer this business question from your data — "What is the average cholesterol level for patients with and without heart disease?"
#Exercise 3: Answer — "Which chest pain type is most common among heart disease patients?"
#Exercise 4: Answer — "What is the age distribution of high-risk vs low-risk patients?"
#Exercise 5: Plot a bar chart showing heart disease count by chest pain type
#Exercise 6: Plot a histogram of age distribution colored by heart disease status
#Exercise 7: Plot a correlation heatmap of all numeric columns

#Ex1
df = pd.read_csv('data/heart.csv')
print(df.isnull().sum())
print(df.dtypes)

#Ex2
avg_cholesterol = df.groupby('HeartDisease')['Cholesterol'].mean()
print(avg_cholesterol)

#Ex3
common_chest_pain = df[df['HeartDisease']==1]['ChestPainType'].mode()[0]
print(common_chest_pain)

#Ex4
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,3))
sns.histplot(data=df, x='Age', bins=30, hue='HeartDisease', kde=True, alpha=0.6)

#Ex5

plt.figure(figsize=(5,3))
sns.countplot(data=df, x='ChestPainType', hue='HeartDisease', palette='Set2')

#Ex6

plt.figure(figsize=(5,3))
sns.histplot(data=df, x='Age', bins=30, hue='HeartDisease', kde=True, alpha=0.6)

#Ex7 
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True , cmap='coolwarm',fmt='.2f', center=0)
```
