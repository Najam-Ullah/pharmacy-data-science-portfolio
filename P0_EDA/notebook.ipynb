```python
import pandas as pd
df = pd.read_csv('data/heart.csv')
print(df.shape)
print(df.head)
print(df.isnull().sum)
print(df.dtypes)
print(df.describe)
```

```python
 # DAY 2: Indexing, Selecting, Assigning ----

col = df['Sex']  
print(type(col))       

# Select multiple columns
subset = df[['Age', 'Sex']]
print(subset.head())

# iloc - integer position based
print(df.iloc[0])        # first row
print(df.iloc[0:5, 0:3]) # first 5 rows, first 3 columns

# loc - label based
print(df.loc[0:4, 'Age':'Sex'])  
```
